```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)

df = pd.read_csv (r'/content/adviseinvest.csv')
```

```python
df.describe()
```

```python
df_clean = df[(df['income'] > 0) & (df['num_accts'] < 3000)]
df_clean = df_clean.dropna()
df_clean.describe()
```

```python
df_clean['product'] = df_clean['product'].astype('category')
df_clean['job'] = df_clean['job'].astype('category')
df_clean['chk_acct'] = df_clean['chk_acct'].astype('category')
df_clean['num_accts'] = df_clean['num_accts'].astype('category')

df_clean.info()
```

```python
answered_mean = df_clean['answered'].mean()
print("Mean of 'answered' after cleaning the dataset:", answered_mean)
```
