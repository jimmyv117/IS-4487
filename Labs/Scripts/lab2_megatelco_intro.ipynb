## Data Visualization

➡️ Assignment Tasks
- Add at least two charts to show a summary of any major trends

```python
df_counts = df['reported_satisfaction'].value_counts()

df_counts.plot(kind='bar', color=['purple', 'orange'])

plt.xlabel('Reported Satisfaction')
plt.ylabel('Count')
plt.title('Reported Satisfaction Bar Chart')

plt.show()
```

```python
df_counts = df['leave'].value_counts()

labels = [f'{label} ({count})' for label, count in zip(['LEAVE','STAY'], df_counts)]

plt.figure(figsize=(8,8))
plt.pie(df_counts,labels=labels,autopct='%1.1f%%',startangle=90)
plt.title('Stay or Leave')
plt.show()
```
