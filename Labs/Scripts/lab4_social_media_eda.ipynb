## Summarize Data

➡️ Assignment Tasks
- Create at least two charts to summarize and visualize the data
- Examples include: Number of posts per device; top 10 posts with highest engagement

```python
iphone_string = '<a href="http://twitter.com/download/iphone" rel="nofollow">Twitter for iPhone</a>'
android_string = '<a href="http://twitter.com/download/android" rel="nofollow">Twitter for Android</a>'

def extract_device_info(s):
    index_for = s.find("for")
    if index_for != -1:
        return s[index_for + len("for "):]
    else:
        return "Unknown"

tweets = [iphone_string, android_string]
platforms = [extract_device_info(tweet) for tweet in tweets]

platform_counts = {}
for platform in platforms:
    platform_counts[platform] = platform_counts.get(platform, 0) + 1

labels = platform_counts.keys()
sizes = platform_counts.values()

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Tweets: iPhone vs. Android')
plt.show()
```

```python
df['reply_count'] = df['reply_count'].astype('float64')
df['retweet_count'] = df['retweet_count'].astype('float64')

total_reply_count = df['reply_count'].sum()
total_retweet_count = df['retweet_count'].sum()

labels = ['Reply Count', 'Retweet Count']
counts = [total_reply_count, total_retweet_count]

plt.bar(labels, counts, color=['skyblue', 'lightcoral'])
plt.title('Total Reply Count vs Total Retweet Count')
plt.xlabel('Counts')
plt.ylabel('Total')
plt.show()
```
